## Gabi Lab One

The data in this lab is from a project run by [Katharine Suding's lab](http://www.colorado.edu/sudinglab/) in the EBIO dept. at the [Niwot Ridge Long Term Ecological Research Site](http://niwot.colorado.edu/). It is a list of the plots and plant species that they sampled across a high elevation talus landscape. At each plot, there could be 1, 2, or 3 species to sample. The goal of the sampling was to take roots from the 3 most abundant plants at each plot. But, since the sampling is destructive (digging up the plant), they did not take any plants that had less than 5 individuals in that plot. Because of this, not all plots have 3 species listed on the data sheet. Species names were listed by their 6 letter species code (the first 3 letters of their genus and first 3 letters of their species name).

Let's import the data and **tidy** it up!

```python
data = pd.read_csv('RootProtocol.csv')
data
```

## What needs to change?

The species need to be organized by plot number so that each observation has a 
**Variable:** *Species Number* and **Value:** *Species Name*

```python
tidy = pd.melt(data, id_vars=['Plot'], var_name='Species Number', value_name='Species Name')
tidy
```

Now we should sort the table by Plot number 

```python
tidy.sort_values(by=['Plot'], inplace=True)
tidy
```

Now the observations are sorted by plot number and are ready for analyzation!
