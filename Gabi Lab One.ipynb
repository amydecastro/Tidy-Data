The goal of the sampling was to take roots from the 3 most abundant plants at each plot. But, since the sampling is destructive (digging up the plant), they did not take any plants that had less than 5 individuals in that plot. Because of this, not all plots have 3 species listed on the data sheet. Species names were listed by their 6 letter species code (the first 3 letters of their genus and first 3 letters of their species name).

Let's import the data and **tidy** it up!

```python
data = pd.read_csv('RootProtocol.csv')
data
```

## What needs to change?

The species need to be organized by plot number so that each observation has a 
**Variable:** *Species Number* and **Value:** *Species Name*

```python
tidy = pd.melt(data, id_vars=['Plot'], var_name='SpeciesNumber', value_name='SpeciesName')
tidy
```

Now we can sort the table by Plot number 

```python
tidy.sort_values(by=['Plot'], inplace=True)
tidy
```

Now let's plot the results!

```python
import matplotlib.pyplot as plt
pd.set_option('max_columns', 50)
%matplotlib inline
```

```python
ax = tidy.groupby("SpeciesName").SpeciesNumber.count().plot(kind="bar")
```

Here we can see that FESBRA (Festuca brachyphylla) is by far the most common plant species that was found in the plots. TRISPI (Trisetum spicatum) was the second most commone plant species found. We can also see that several species were only found in one plot. The next step with this data would be to count the number of plants of each species found in each plot.
